```python
import pandas as pd
import numpy as np
import pickle
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load embeddings
layer_embeddings = pickle.load(open("../../code/gnn/product_space_embeddings.pickle", "rb"))
# Commodity classes
product_codes = [f"{x:02}" for x in range(1,100) if x not in [77, 98, 99]]
```

```python
layer_embeddings.keys()
```

```python
# Choose a year
year = 2016
```

```python
similarity_matrix = cosine_similarity(layer_embeddings[year].detach().numpy())

# Create a DataFrame for the similarity matrix for better labeling in the heatmap.
sim_df = pd.DataFrame(similarity_matrix, index=product_codes, columns=product_codes)
```

```python
# Plot the Heatmap using Seaborn
fig, ax = plt.subplots(figsize=(20, 18), dpi=600)
heatmap = sns.heatmap(sim_df, cmap='viridis', xticklabels=True, yticklabels=True, ax=ax, cbar_kws={"shrink": 1, "aspect": 40})
#plt.title("Pairwise Cosine Similarity Heatmap of Layer Embeddings")
ax.set_xlabel("Product Code", fontsize=14, weight='bold')
ax.set_ylabel("Product Code", fontsize=14, weight='bold')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
